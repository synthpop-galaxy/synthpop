```python
import synthpop 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pdb
from astropy.table import vstack
```

```python
mod2=0
mod2 = synthpop.SynthPop('huston2025_defaults.synthpop_conf',
                         extinction_map_kwargs={'name':'maps_from_dustmaps', 
                                                'dustmap_name': 'marshall'},
                         chosen_bands = ['ubv', 'ukirt'],
                         maglim = ['m_ukirt_H', 99, "keep"],
                         star_generator="SpiseaGenerator",
                         evolution_class={"name":"SpiseaCluster", "block_spisea_prints":True},
                         ifmr_kwargs={"name":"SpiseaIfmr", "spisea_ifmr_name":"IFMR_N20_Sukhbold"},
                         multiplicity_kwargs={"name":"SpiseaMultiplicity", "spisea_multiplicity_name":"MultiplicityResolvedDK"},
                         post_processing_kwargs=None, #[{"name": "CombineTables"}],
                         name_for_output='mod2test',
                         combine_system_mags = True,
                         opt_iso_props=["log_L", "log_Teff", "log_g","log_R", "phase"],
                        )
```

```python
mod2.init_populations()
```

```python
systems, companions = mod2.process_location(l_deg=3, b_deg=-1, field_shape='circle', field_scale=0.003, field_scale_unit='deg')
```
